```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.metrics import f1_score
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
train_df= pd.read_csv('part2_processed_train.csv')
train_df.shape
```

```python
test_df = pd.read_csv('part2_processed_test.csv')
test_df.shape
```

```python
train_df
```

```python
x=test_df['response_id']
test_df.drop(['response_id'],axis=1,inplace=True)
y=train_df['exit_status']
train_df.drop(['exit_status'],axis=1,inplace=True)
```

```python

# scaler = StandardScaler()
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(train_df)
X_test_scaled = scaler.fit_transform(test_df)
```

```python
model = LogisticRegression()
```

```python

param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],  # Different types of regularization
    'solver': ['liblinear', 'saga', 'lbfgs'],      # Include solvers for small and large datasets
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],    # Wide range for regularization strength
    'l1_ratio': [0.1, 0.5, 0.9],                  # ElasticNet mixing parameter (only for saga)
    'max_iter': [100, 500, 1000]                  # Iterations to ensure convergence
}

```

```python
grid_search = RandomizedSearchCV(n_iter=100,estimator=model, param_distributions=param_grid, cv=10, scoring='f1',n_jobs=-1,verbose=0)
```

```python
grid_search.fit(X_train_scaled, y)
```

```python
final = grid_search.predict(X_test_scaled)
```

```python
predictions = ["Stayed" if pred == 1 else "Left" for pred in final]
```

```python
output_df = pd.DataFrame({
    'response_id': x,
    'Predictions': predictions
})
```

```python
output_df.to_csv('../output/logistic_output.csv', index=False)
```

```python
from sklearn.metrics import f1_score
print("The training F1_score is ",f1_score(grid_search.predict(X_train_scaled),y))
```
