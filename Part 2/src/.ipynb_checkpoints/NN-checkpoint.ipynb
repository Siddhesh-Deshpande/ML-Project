```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('preprocessed_test_new.csv')
train = pd.read_csv('preprocessed_train_new.csv')
```

```python
y_train = train['exit_status']
x_train = train.drop('exit_status', axis = 1)
x_train = x_train.drop('response_id', axis = 1)
x_train.head(5)
```

```python
test.head(5)
```

```python
res = test['response_id']
test = test.drop('response_id', axis = 1)
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
```

```python
x_train.shape
```

```python
test.shape
```

```python
x_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_train)
```

```python
ann = models.Sequential([
    layers.Input(shape=(22, )),  # Correctly specify input shape
    layers.Flatten(),
    layers.Dense(64, activation='relu'),   # Second hidden layer with 64 neurons and ReLU activation
    layers.Dense(32, activation='relu'),   # Third hidden layer with 32 neurons and ReLU activation
    layers.Dense(2, activation='softmax')
])

ann.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['f1_score']
)

# Assuming x_train and y_train are defined
ann.fit(x_scaled, y_train, epochs=10)
```

```python
test_scaled = scaler.fit_transform(test)
y_pred = ann.predict(test_scaled)
```

```python
y_pred
```

```python
y_pred.shape
```

```python
pred = []
for i in range(0, len(y_pred)):
    if y_pred[i][0] > 0.5:
        pred.append('Left')
    else:
        pred.append('Stayed')
pred[:10]
```

```python
data = {
    'response_id': res,
    'Predictions': pred
}
df = pd.DataFrame(data)
```

```python
df.to_csv('../output/NN_pred_1.csv', index = False)
```
